```python
import pandas as pd

data=pd.read_csv(r"/content/penguins.csv")
```

```python
for i in data.columns:
    if data[i].dtype=="O":
        mode=data[i].mode()[0]
        data[i].fillna(mode,inplace=True)
    else:
        mean=data[i].mean()
        data[i].fillna(mean,inplace=True)
```

```python
(data.isnull().sum()*100)/len(data)
```

```python
data.head()
# Change column names bill_length(mm)	bill_depth_mm	flipper_length_mm	body_mass_g
data.rename(columns={
    'bill_length_mm':'bill_length(mm)',
    'bill_depth_mm':'bill_depth(mm)',
    'flipper_length_mm':'flipper_length(mm)',
    'body_mass_g':'body_mass(g)'
})
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

Y=le.fit_transform(data['species'])
Y
```

```python
pd.get_dummies(data)
```

```python
from sklearn.preprocessing import OneHotEncoder
# use for loop
ohe=OneHotEncoder(sparse=False)
c=data.island.unique()
c.sort()
x=pd.DataFrame(ohe.fit_transform(data[['island']]),columns=c)
x
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse=False)
c=data.sex.unique()
c.sort()
y=pd.DataFrame(ohe.fit_transform(data[['sex']]),columns=c)
y
```

```python
X=pd.concat((data,x,y),join='outer',axis=1)
X
d=['species','island','sex']
X1=X.drop(d,axis=1)
X1
```

```python
# for standardlize values
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_scl=pd.DataFrame(sc.fit_transform(X1),columns=X1.columns)
X_scl
```

```python
# to merger featuring engneering
X_scl['bill_volume']=X_scl['bill_length_mm']*X_scl['bill_depth_mm']
```

```python
# to select feature selection we use decision tree
X_scl.head()
```

```python
Xf=X_scl.drop(['bill_length_mm','bill_depth_mm'],axis=1)
Xf
```

```python
Y
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree

dtc=DecisionTreeClassifier()
dtc.fit(Xf,Y)
```

```python
dtc.feature_importances_
```

```python
sp=['Biscoe', 'Dream', 'Torgersen']
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(dpi=300)
plot_tree(dtc,class_names=sp,feature_names=Xf.columns)
plt.show()
```
