```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("https://raw.githubusercontent.com/rashida048/Datasets/master/fraud_data.csv ")
```

```python
data
```

```python
data.shape
```

```python
len(data['V1'].unique())
```

```python
# dropping the case number columns as it is not required
data.drop(["V1"],inplace=True,axis = 1)
```

```python
#shape of dataset
data.shape
```

```python
data.head()
```

```python
# Removing NA values in data set
data = data.dropna()
data.shape
```

```python
# Dividing our data into input and output variables 
X = data.iloc[:,:28]
Y = data.iloc[:,28]
```

```python
X
```

```python
Y
```

```python
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression()
classifier.fit(X,Y)

# classifier.write_to_pickle('path of file.pkl')
# classifier.save('Model.hd5')
```

```python
# save the model to disk
filename = 'finalized_model.sav'
pickle.dump(classifier, open(filename, 'wb'))
```

```python
#Predict for X dataset
# pickle.load(open(filename, 'rb'))
# classifier.read_pickle_file('/content/finalized_model.sav')
y_pred = classifier.predict(X)
```

```python
y_pred_df= pd.DataFrame({'actual': Y,
                         'predicted_prob': classifier.predict(X)})
```

```python
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
print (confusion_matrix)
```

# ROC Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, classifier.predict_proba (X)[:,1])

auc = roc_auc_score(Y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.legend()
```
