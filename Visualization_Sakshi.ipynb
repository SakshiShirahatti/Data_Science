```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
%matplotlib inline
```

```python
x=np.arange(1,201,2)
y=np.random.randint(1,10,100)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(15,5))
plt.axes(facecolor="pink")
plt.plot(x,y,color="black",label="Line Chart")
plt.xlim(0,200)
plt.ylim(0,10)
plt.xlabel("Sales")
plt.ylabel("Product")
plt.title("Line Plot")
plt.suptitle("Sales Comparision",size=20,y=1.05)
plt.savefig("Save.png")
plt.legend()
plt.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0,0].set_title('Plot 1')
ax[0,0].plot(x,y)
plt.tight_layout()
```

```python
fig, ax=plt.subplots(nrows=2,ncols=2)
ax[0,1].plot(x,y)
plt.tight_layout()
```

```python
fig, ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].plot(x,y)
ax[0,1].plot(x,y)
ax[1,0].plot(x,y)
ax[1,1].plot(x,y)
plt.tight_layout()
```

```python
fig,ax=plt.subplots(nrows=4,ncols=3,figsize=(10,10))
ax[0,0].plot(x,y,color="black",linestyle="dashed")
ax[0,1].scatter(x,y)
ax[0,2].plot(x,y,color="yellow")
ax[1,0].plot(x,y,color="green")
ax[1,1].plot(x,y)
ax[1,2].plot(x,y)
ax[2,0].plot(x,y)
ax[2,1].plot(x,y)
ax[2,2].plot(x,y)
ax[3,0].plot(x,y)
ax[3,1].plot(x,y)
ax[3,2].plot(x,y)
plt.show()
```
