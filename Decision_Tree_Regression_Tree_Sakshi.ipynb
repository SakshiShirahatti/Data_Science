# **Decision Tree**

```python
from sklearn.datasets import load_breast_cancer
```

```python
# model training
x=load_breast_cancer().data
y=load_breast_cancer().target

# for visualization
fn=load_breast_cancer().feature_names
cn=load_breast_cancer().target_names
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
dtc=DecisionTreeClassifier(criterion="entropy")
```

```python
dtc.fit(x_train,y_train)
```

```python
pred=dtc.predict(x_test)
pred
```

```python
import matplotlib.pyplot as plt
fix,ax=plt.subplots(dpi=500)
#plt.Figure(dpi=3000)
plot_tree(dtc,feature_names=fn,class_names=cn)
plt.show()
```

# **Regression Tree**

```python
from sklearn.datasets import fetch_california_housing
```

```python
fetch_california_housing()
```

```python
x=fetch_california_housing().data
y=fetch_california_housing().target
# fn=fetch_california_housing().feature_name
# cn=fetch_california_housing().target_name
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor(min_samples_split=6)
```

```python
dtr.fit(x_train,y_train)
```

```python
dtr.score(x_test,y_test)
```
